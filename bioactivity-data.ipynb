```python
#Importing Libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

**Target search for Corona Virus**

```python
#To search for the target Protein of Coronavirus
target = new_client.target
target_query = target.search("coronavirus")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#Retrieving and selecting bioactivity data for SARS coronavirus 3C-like proteinase(5th entry)
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
#to retrieve boactivity for ARS coronavirus 3C-like proteinase('CHEMBL3927')
activity = new_client.activity
res= activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

**Handling Missing Data**

```python
df2 = df.dropna(axis = 0, how = "all")
df2
```

**Data pre-processing of the bioactivity data**                                 

The bioactivity data in the IC50 unit compounds having values of less than 1,000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. Values between 1,000 nM will be refered to as Intermediate

```python
bioactivity_class = []
for i in df.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```
